```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd
import re
import ast
import unidecode
import time
import numpy as np
import random
from random import randint

resultados=np.array(('pais','categoria','tematica','tecnica','soporte','medidas','fecha','etiquetas','precio','obra','autor','link'))
```

```python
for i in range(1,66):
    url="https://www.artelista.com/pintura-seleccion-artelista-precio-mas-de-4000euros-T1P0E205V1-0-0Lall-p"+str(i)+".html"
    r=requests.get(url)
    print('pagina:', i)
    soup = BeautifulSoup(r.text)
    links=[]
    for a in soup.find_all('a', href=True):
        links.append(a['href'])

    subs = '/obra/'
    links = list(filter(lambda x: subs in x, links))
    links = list(set(links))
    for link in links:
        time.sleep(randint(1,2))
        r=requests.get(link)
        req=unidecode.unidecode(r.text)
        soup = BeautifulSoup(req,'html.parser')
        div_bs4 = soup.findAll("div", {"id": "info_obra"})
        div_precio = soup.findAll("div",{"itemprop":"offers"})
        div_nombre = soup.find("div",{"id":"obra"})
        features=[]
        for res in div_bs4:
            features.append(res.findAll(['li']))
        clean_features=[]
        for feature in features[0]:
            clean_features.append(feature.text.partition(" ")[1:][1])
        if (len(clean_features)<8):
            clean_features.append('sin etiquetas')
        if len(div_precio)==0:
            clean_features.append('Obra vendida')
        else:
            clean_features.append(div_precio[0].find(itemprop="price").text)
        nombre_obra=div_nombre.find(itemprop="name").text
        autor=div_nombre.find(itemprop="author").text
        clean_features.append(nombre_obra)
        clean_features.append(autor)
        clean_features.append(link)
        clean_features=np.array(clean_features)
        if (len(clean_features)==12):
            resultados = np.vstack((resultados, clean_features))
            print(nombre_obra)
        else:
            pass
        if (resultados.shape[0] % 50 == 0):
            print("guardando...")
            pd.DataFrame(resultados[1:], columns=resultados[0]).to_csv(f'/Users/guillermonaranjomuedano/Documents/Escuela/Materias Antiguas/7° Semestre/Tópico de Negocios II/Foncarte/datasets/guardado_parcial.csv', header=True, index=False)
pd.DataFrame(resultados[1:], columns=resultados[0]).to_csv(f'/Users/guillermonaranjomuedano/Documents/Escuela/Materias Antiguas/7° Semestre/Tópico de Negocios II/Foncarte/datasets/dataset_artelista.csv', header=True, index=False)
```
