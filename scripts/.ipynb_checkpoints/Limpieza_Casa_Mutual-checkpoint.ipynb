```python
import requests
```

```python
#aqui primero descargamos el csv desde el github que tenemos
csv_url='https://raw.githubusercontent.com/MarianoAlcarazAguilar/Foncarte/main/datasets/data_mutualart_completo.csv'
req = requests.get(csv_url)
url_content = req.content
csv_file = open('data_mutualart_completo.csv', 'wb')
csv_file.write(url_content)
csv_file.close()
```

```python
import pandas as pd
df=pd.read_csv('data_mutualart_completo.csv')
```

```python
df
```

```python
df.reset_index()
```

```python
casa_subasta=df['casa_subasta'].tolist()
```

```python
#te entrega los unique values de una lista
def unique(list1):
  
    # initialize a null list
    unique_list = []
  
    # traverse for all elements
    for x in list1:
        # check if exists in unique_list or not
        if x not in unique_list:
            unique_list.append(x)
    
    return unique_list
    # print list
    #for x in unique_list:
     #   print(x)
```

```python
unicos=unique(casa_subasta) #esto nos va ayudar a identificar los nombres que se repiten 
sorted(unicos)
```

```python
#hacemos una lista con version oficial de los nombres que se repiten (simon fue a manita)
nomPorLimpiar=['33 Auction','Azur','Bonhams','Bruun Rasmussen','Cambi Auction House','Christie\'s','Cornette de Saint Cyr',
               'DVC','Dreweatts','Fine Art Auctions','Heffel','Ketterer Kunst','Koller','Lyon & Turnbull','Phillips',
              'Pierre Bergé & Associates','Sotheby\'s']
```

```python
for i in range(len(nomPorLimpiar)):
    for j in range(len(casa_subasta)):
        if nomPorLimpiar[i] in casa_subasta[j]:
            casa_subasta[j]=nomPorLimpiar[i]
```

```python
df['casa_subasta']=casa_subasta
```

```python
df.to_csv('../data_mutualart_completo.csv',header=True,index=False)
```
