```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas_datareader as web
```

```python
gold = web.get_data_yahoo("GLD",
                            start = "2000-01-01")
```

```python
gold
```

```python
gold=gold.drop(['Low', 'High','Open','Volume','Close'], axis=1)
```

```python
gold['Adj Close'].plot()
plt.xlabel("Date")
plt.ylabel("Adjusted")
plt.title("Gold Price data")
plt.show()
```

```python
daily_returns = gold['Adj Close'].pct_change()
```

```python
daily_returns.tolist()
```

```python
gold["daily_return"]=daily_returns.tolist()
```

```python
gold
```

```python
gold['daily_return'].plot(title='gold daily returns')
```

```python
gold['Adj Close'].plot(title='Gold Price')
```

```python
gold.reset_index(inplace=True)
```

```python
gold.to_csv('../gold.csv',header=True,index=False)
```
