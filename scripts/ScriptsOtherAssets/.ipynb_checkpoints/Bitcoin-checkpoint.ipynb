```python
new = HistoricalData('BTC-USD',86400,'2000-01-01-00-00').retrieve_data()
```

```python
new
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas_datareader as web
```

```python
close=new.drop(['low', 'high','open','volume'], axis=1)
```

```python
close
```

```python
daily_returns = new['close'].pct_change()
```

```python
daily_returns.tolist()
```

```python
close["daily_return"]=daily_returns.tolist()
```

```python
bitcoin=close
```

```python
bitcoin
```

```python
bitcoin['daily_return'].plot(title='Bitcoin daily returns')
```

```python
#evolucion en el precio
bitcoin['close'].plot(title='Bitcoin Price')
```
