```python
import requests
```

```python
#aqui primero descargamos el csv desde el github que tenemos
csv_url='https://raw.githubusercontent.com/MarianoAlcarazAguilar/Foncarte/main/datasets/data_mutualart_completo.csv'
req = requests.get(csv_url)
url_content = req.content
csv_file = open('data_mutualart_completo.csv', 'wb')
csv_file.write(url_content)
csv_file.close()
```

```python
import pandas as pd
df=pd.read_csv('data_mutualart_completo.csv')
```

```python
df
```

```python
df['casa_subasta'] = df['casa_subasta'].str.split(',').str[0]
```

```python
df['casa_subasta'] = df['casa_subasta'].str.split('-').str[0]
```

```python
df["casa_subasta"] = df["casa_subasta"].str.replace(',','')
```

```python
df["casa_subasta"] = df["casa_subasta"].str.replace('  ','')
```

```python
df['casa_subasta'] = df['casa_subasta'].str.split('\'s').str[0]
```

```python
df['casa_subasta'] = df['casa_subasta'].str.split('´s').str[0]
```

```python
df.reset_index()
```

```python
df.to_csv('../data_mutualart_completo.csv',header=True,index=False)
```

```python
#te entrega los unique values de una lista
def unique(list1):
  
    # initialize a null list
    unique_list = []
  
    # traverse for all elements
    for x in list1:
        # check if exists in unique_list or not
        if x not in unique_list:
            unique_list.append(x)
    # print list
    for x in unique_list:
        print(x)
```
